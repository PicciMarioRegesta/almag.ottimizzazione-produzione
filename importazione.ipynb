```python
import pandas as pd
import numpy as np
```

```python
input_filepath = 'Dati.xlsx'
```

## Produzioni

Per cartellino (ricetta)

```python
prod_df_raw = pd.read_excel(
    input_filepath,
    sheet_name="Produzioni"
)
prod_df_raw.head()
```

```python
prod_df = prod_df_raw[['Cartellino', 'Quantita']].groupby('Cartellino').sum().sort_values(['Cartellino']).reset_index()
prod_df.head()
```

```python
produzioni = np.array(prod_df['Quantita'])
print(produzioni.shape)
produzioni
```

## Consumi

Per famiglia

```python
cons_df_raw = pd.read_excel(
    input_filepath,
    sheet_name="Consumi"
)
cons_df_raw.head()
```

```python
cons_df = cons_df_raw[['Famiglia', 'Quantità']].groupby('Famiglia').sum().sort_values(['Quantità']).sort_values(['Famiglia']).reset_index()
cons_df['Quantità'] = cons_df['Quantità']*-1
cons_df.head()
```

```python
consumi = np.array(cons_df['Quantità'])
print(consumi.shape)
consumi
```

## Cartellino standard (ricetta)

```python
cart_df_raw = pd.read_excel(
    input_filepath,
    sheet_name="Cartellino Standard"
)

cart_df_raw = cart_df_raw[cart_df_raw['Cartellino'].isin(prod_df['Cartellino'])]
cart_df_raw = cart_df_raw[cart_df_raw['Famiglia'].isin(cons_df['Famiglia'])]

cart_df_raw.head()
```

```python
cart_df = cart_df_raw[['Cartellino', 'Famiglia', 'Percentuale']].reset_index().pivot_table(
    index=['Famiglia'],
    columns=['Cartellino'],
    values='Percentuale'
).fillna(0)

print(cart_df.shape)
cart_df
```

```python
ricette = np.array(cart_df).flatten()
```
