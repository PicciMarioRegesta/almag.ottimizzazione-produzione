```python
import pandas as pd
import numpy as np
```

```python
input_filepath = 'Dati.xlsx'
```

## Produzioni

Per cartellino (ricetta)

```python
prod_df_raw = pd.read_excel(
    input_filepath,
    sheet_name="Produzioni"
)
prod_df_raw.head()
```

```python
prod_df = prod_df_raw[['Cartellino', 'Quantita']].groupby('Cartellino').sum().sort_values(['Cartellino'])
prod_df

# prod_df.to_clipboard(excel=True,sep='\t',  decimal=',')
```

```python
produzioni = np.array(prod_df['Quantita'])
print(produzioni.shape)
produzioni
```

## Consumi

Per famiglia

```python
cons_df_raw = pd.read_excel(
    input_filepath,
    sheet_name="Consumi"
)
cons_df_raw.head()
```

```python
cons_df = cons_df_raw[['Famiglia', 'Quantità']].groupby('Famiglia').sum().sort_values(['Quantità']).sort_values(['Famiglia'])
cons_df['Quantità'] = cons_df['Quantità']*-1
cons_df

# cons_df.to_clipboard(excel=True,sep='\t',  decimal=',')
```

```python
consumi = np.array(cons_df['Quantità'])
print(consumi.shape)
np.sum(consumi)
consumi
```

## Cartellino standard (ricetta)

```python
cart_df_raw = pd.read_excel(
    input_filepath,
    sheet_name="Cartellino STANDARD Partenza",
    # sheet_name="Cartellino consuntivo",
    # header=1
)

cart_df_raw = cart_df_raw[cart_df_raw['Cartellino'].isin(prod_df.index)]
cart_df_raw = cart_df_raw[cart_df_raw['Famiglia'].isin(cons_df.index)]

cart_df_raw
```

```python
cart_df = cart_df_raw[['Cartellino', 'Famiglia', 'Percentuale']].reset_index().pivot_table(
    index=['Famiglia'],
    columns=['Cartellino'],
    values='Percentuale'
).fillna(0)

print(cart_df.shape)
cart_df

# cart_df.to_clipboard(excel=True,sep='\t',  decimal=',')
```

```python
ricette = np.array(cart_df).flatten()/100
ricette
```

## Valore atteso materiale per ricetta

```python
att_raw = pd.read_excel(
    input_filepath,
    sheet_name="Toll Fonderia Medio",
    header=1,
    usecols='A,AC:AO'
)
att_raw = att_raw.loc[att_raw['lega'].isin(prod_df.index)]
att_raw = att_raw.set_index('lega')
print(att_raw.shape)
att_raw
```

```python
a = np.array(att_raw.reset_index()['lega'])
b = np.array(prod_df.index).astype(int)

for i in range(len(b)):
    if (b[i] not in a):
        print(b[i])
        att_raw = att_raw.reindex(att_raw.index.values.tolist()+[b[i]]).fillna(0).sort_index()

att_raw
```

```python
val_att = np.array(att_raw)
print(val_att.shape)
val_att
```

```python
val_att

range_ric_mat = np.full((val_att.shape[0],val_att.shape[1]), None, dtype='f,f')

for iy, ix in np.ndindex(val_att.shape):
    range_ric_mat[iy, ix] = (val_att[iy, ix]/100 if val_att[iy,ix] != 0 else 0, 0.01)

print(range_ric_mat.shape)
range_ric_mat
```
